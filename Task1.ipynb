```python
# Task 1: Iris Dataset Exploration
# AI/ML Internship - DevelopersHub Corporation

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
iris = sns.load_dataset('iris')

# 1. Data Inspection
print("="*50)
print("DATASET INSPECTION")
print("="*50)
print(f"Shape: {iris.shape}")
print(f"\nColumns: {list(iris.columns)}")
print("\nFirst 5 rows:")
print(iris.head())
print("\nInfo:")
print(iris.info())
print("\nStatistical Summary:")
print(iris.describe())

# 2. Visualizations
plt.style.use('seaborn-v0_8-darkgrid')

# Scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=iris, x='petal_length', y='petal_width', hue='species', s=100)
plt.title('Petal Length vs Petal Width', fontsize=14)
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.legend(title='Species')
plt.show()

# Histograms
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

for i, feature in enumerate(features):
    row, col = i//2, i%2
    sns.histplot(data=iris, x=feature, hue='species', kde=True, ax=axes[row, col])
    axes[row, col].set_title(f'Distribution of {feature.replace("_", " ").title()}')
    axes[row, col].set_xlabel(feature.replace("_", " ").title())

plt.tight_layout()
plt.show()

# Box plots
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

for i, feature in enumerate(features):
    row, col = i//2, i%2
    sns.boxplot(data=iris, x='species', y=feature, ax=axes[row, col])
    axes[row, col].set_title(f'Box Plot: {feature.replace("_", " ").title()} by Species')
    axes[row, col].set_xlabel('Species')
    axes[row, col].set_ylabel(feature.replace("_", " ").title())

plt.tight_layout()
plt.show()

# Pairplot (additional visualization)
sns.pairplot(iris, hue='species', diag_kind='kde', height=2.5)
plt.suptitle('Pairplot of Iris Dataset', y=1.02, fontsize=14)
plt.show()

# 3. Observations
print("\n" + "="*50)
print("OBSERVATIONS")
print("="*50)
print("""
1. Dataset contains 150 samples with 4 features and 1 target (species).
2. No missing values detected in any column.
3. Petal length and petal width show clear separation between all three species.
4. Setosa species is completely distinct from versicolor and virginica.
5. Sepal width shows some overlap, with possible mild outliers.
6. Versicolor and virginica have some overlap in sepal measurements.
7. All features are on a similar scale (cm), so no scaling needed initially.
""")
```
